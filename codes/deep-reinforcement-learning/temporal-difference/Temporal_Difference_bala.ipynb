Your algorithm has four arguments:

````python
EPSILON_START = 1.0
EPSILON_DECAY = 0.9999
EPSILON_MIN = 0.1
Gamma = 1.0 #Discount

def Epsilon_GreedyPolicy(Q, state, epsilon, nA):
    rand = np.random.uniform(0,1)
    if rand < epsilon:
        return np.random.choice( np.arange(nA) )
    else:
        return np.argmax(Q[state])
    
def UpdateQ(Q, state, action, reward, next_state, next_action, alpha):
    current_QA = Q[state][action]
    Q_Next = Q[next_state][next_action] #if next_state is not None else 0
    target = reward +  (Gamma* Q_Next )
    return current_QA + alpha*(target - current_QA)

def sarsa(env, num_episodes, alpha, gamma=1.0):
    
    # initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    # initialize performance monitor
    # loop over episodes
    epsilon = EPSILON_START
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()   
        
        ## TODO: complete the function
        
        # Observe the state 
        state = env.reset()
        
        # Set epsilon <- epsilon_i
        epsilon = 1.0 / i_episode #max( epsilon * EPSILON_DECAY, EPSILON_MIN)
       
        # Choose an action
        action = Epsilon_GreedyPolicy(Q, state, epsilon, env.nA)
        
        while True:

            # Take action    
            next_state, reward, done, info = env.step(action)
            
            #if not done:
            #Choose action At+1
            next_action = Epsilon_GreedyPolicy(Q, next_state, epsilon, env.nA)

            Q[state][action] = UpdateQ(Q, state, action, reward, next_state, next_action, alpha)
            state = next_state
            action = next_action

            if done:
                #Q[state][action] = UpdateQ(Q, state, action, reward, None, None, alpha)
                break
        
    return Q
````

### Part 2: TD Control: Q-learning

In this section, you will write your own implementation of the Q-learning control algorithm.

Your algorithm has four arguments:

- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
EPSILON_START = 1.0
EPSILON_DECAY = 0.9999
EPSILON_MIN = 0.1
Gamma = 1.0 #Discount

def GreedyPolicy(Q, state, epsilon, nA):
    return np.argmax(Q[state])

def Epsilon_GreedyPolicy(Q, state, epsilon, nA):
    rand = np.random.uniform(0,1)
    if rand < epsilon:
        return np.random.choice( np.arange(nA) )
    else:
        return np.argmax(Q[state])
    
def UpdateQ_SarsaMax(Q, state, action, reward, next_state, alpha):
    current_QA = Q[state][action]
    Q_max = np.max(Q[next_state])
    target = reward +  (Gamma* Q_max )
    return current_QA + alpha*(target - current_QA)


def q_learning(env, num_episodes, alpha, gamma=1.0):
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(env.nA))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODO: complete the function
        # Observe the state 
        state = env.reset()
        
        # Set epsilon <- epsilon_i
        epsilon = 1.0 / i_episode #max( epsilon * EPSILON_DECAY, EPSILON_MIN)
               
        while True:

            # Choose an action
            action = Epsilon_GreedyPolicy(Q, state, epsilon, env.nA)
            
            # Take action    
            next_state, reward, done, info = env.step(action)
            
            Q[state][action] = UpdateQ_SarsaMax(Q, state, action, reward, next_state, alpha)
            state = next_state

            if done:
                break
        
    return Q
````

### Part 3: TD Control: Expected Sarsa

In this section, you will write your own implementation of the Expected Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
EPSILON_START = 1.0
EPSILON_DECAY = 0.9999
EPSILON_MIN = 0.1
Gamma = 1.0 #Discount

def GreedyPolicy(Q, state, epsilon, nA):
    return np.argmax(Q[state])

def Epsilon_GreedyPolicy(Q, state, epsilon, nA):
    rand = np.random.uniform(0,1)
    if rand < epsilon:
        return np.random.choice( np.arange(nA) )
    else:
        return np.argmax(Q[state])
    
def get_probs(Q_s, epsilon, nA):
    """ obtains the action probabilities corresponding to epsilon-greedy policy """
    policy_s = np.ones(nA) * epsilon / nA
    best_a = np.argmax(Q_s)
    policy_s[best_a] = 1 - epsilon + (epsilon / nA)
    return policy_s

def UpdateQ_ExpectedSarsaMax(Q, state, action, reward, next_state, alpha, epsilon, nA):
    current_QA = Q[state][action]
    pi_a_given_nextstate = get_probs(Q[next_state], epsilon, nA)
    Expected_Sarsa_max = np.dot(Q[next_state] , pi_a_given_nextstate )
    target = reward +  (Gamma* Expected_Sarsa_max )
    return current_QA + alpha*(target - current_QA)


def expected_sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(env.nA))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODO: complete the function
        # Observe the state 
        state = env.reset()
        
        # Set epsilon <- epsilon_i
        epsilon = 0.005 # 1.0 / i_episode #max( epsilon * EPSILON_DECAY, EPSILON_MIN)
               
        while True:

            # Choose an action
            action = Epsilon_GreedyPolicy(Q, state, epsilon, env.nA)
            
            # Take action    
            next_state, reward, done, info = env.step(action)
            
            Q[state][action] = UpdateQ_ExpectedSarsaMax(Q, state, action, reward, next_state, alpha, epsilon, env.nA)
            state = next_state

            if done:
                break
    return Q
````
